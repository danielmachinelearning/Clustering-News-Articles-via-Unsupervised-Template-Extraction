```python
## Convert the overall counts into dataframes.
df1 = pd.DataFrame(overallVerbCounts)
df2 = pd.DataFrame(overallDepCounts)
df3 = pd.DataFrame(overallTokenPairs)
df4 = pd.DataFrame(overallPOS_tag_counts)
df5 = pd.DataFrame(overallArgPairs)
df6 = pd.DataFrame(overallEntCounts)
df7 = pd.DataFrame(overallResults)
df8 = pd.DataFrame(overallAverageVectors)

## Combine all dataframes columnwise into one dataframe.
horizontal_stack = pd.concat([df1, df2, df3, df4, df5, df6, df7, df8], axis=1)
horizontal_stack = horizontal_stack.drop(horizontal_stack.columns[0], axis=1)
horizontal_stack = horizontal_stack.fillna(0)
print(horizontal_stack)
```

```python
from sklearn.cluster import AgglomerativeClustering
from sklearn.preprocessing import StandardScaler
import umap

scaler = StandardScaler()
X_std = scaler.fit_transform(horizontal_stack)
embedding = umap.UMAP(n_components=13).fit_transform(X_std)

model = AgglomerativeClustering(n_clusters=4, affinity='cosine', compute_full_tree='auto', linkage='complete')
model.fit_predict(embedding)
print(model.labels_)
```
